### Analysis

This study result will primarily revolve around the distribution of tumor volumes  and which drugs had tangible effects throughout the study. In our boxplot of Distribution of Tumor volume for each treatment group, we can see Capomulin and Ramicane having noticeably lower interquartile ranges than that of Infubinol and Ceftamin.  This is a very obvious sign that Capomulin and Ramicane both have had an affect on the tumor volumes of their respective mice. 

Our line plot is another graph depicting the success Capomulin shown throughout the study. Within cell 17, we can see a displayed line graph showing every Mouse’s tumor volume’s over the course of the study. We can see a noticeable trend downward in tumor volume. Our selected Mouse ID, s185, if selected by picking the first Id available from our filtered merged data frame, further filtered by drug regime being Capomulin. I.e. the larger the tumor volume, the more likely mouse weight will increase as well, and vice versa. 

 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import pearsonr, linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_df = pd.merge(mouse_metadata, study_results, on='Mouse ID')

# Display the data table for preview
merged_df
```

```python
# Checking the number of mice.
merged_df['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = merged_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)

duplicate_mice_ids = merged_df[duplicate_mice]['Mouse ID'].unique()

print(duplicate_mice_ids)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_data = merged_df[duplicate_mice]

duplicate_mice_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_merged_df = merged_df[merged_df['Mouse ID'].isin(duplicate_mice_ids) == False]
clean_merged_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_merged_df['Mouse ID'].nunique()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Filter the DataFrame for the specified treatment regimens
treatment_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
filtered_df = merged_df[merged_df['Drug Regimen'].isin(treatment_regimens)]


# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = filtered_df.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume = pd.merge(last_timepoint, filtered_df, on=['Mouse ID', 'Timepoint'])

# Display the final tumor volume for each mouse
final_tumor_volume[['Mouse ID', 'Drug Regimen', 'Tumor Volume (mm3)']]
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Loop through each treatment
for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volumes = final_tumor_volume.loc[final_tumor_volume['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']
    
    # Add the tumor volumes to the list
    tumor_vol_data.append(tumor_volumes)
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = tumor_volumes.quantile([0.25, 0.5, 0.75])
    lower_bound = quartiles[0.25] - 1.5 * (quartiles[0.75] - quartiles[0.25])
    upper_bound = quartiles[0.75] + 1.5 * (quartiles[0.75] - quartiles[0.25])
    
    # Determine outliers using upper and lower bounds
    outliers = tumor_volumes.loc[(tumor_volumes < lower_bound) | (tumor_volumes > upper_bound)]
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
# Create a box plot

plt.boxplot(tumor_vol_data, labels=treatments)

# Set the plot title and y-axis label
plt.title('Distribution of Tumor Volume for Each Treatment Group')
plt.ylabel('Tumor Volume (mm3)')

# Display the plot with grid lines
plt.grid(True)
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# Filter the DataFrame to include only data for the Capomulin regimen
capomulin_df = filtered_df[filtered_df['Drug Regimen'] == 'Capomulin']

# Calculate the average observed tumor volume for each mouse in the Capomulin regimen
average_tumor_volume = capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Merge the average tumor volume data with the 'Weight (g)' column of the DataFrame
merged_df = pd.merge(capomulin_df[['Mouse ID', 'Weight (g)']].drop_duplicates(), average_tumor_volume, on='Mouse ID')

# Create a scatter plot of mouse weight vs. average observed tumor volume
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.scatter(merged_df['Weight (g)'], merged_df['Tumor Volume (mm3)'], marker='o', color='b', edgecolors='black')

# Set the plot title and axis labels
plt.title('Mouse Weight vs. Average Observed Tumor Volume for Capomulin Regimen')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Display the plot
plt.grid(True)  # Optional: Add grid lines
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen



# Calculate the average observed tumor volume for each mouse in the Capomulin regimen
average_tumor_volume = capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Merge the average tumor volume data with the 'Weight (g)' column of the DataFrame
merged_df = pd.merge(capomulin_df[['Mouse ID', 'Weight (g)']].drop_duplicates(), average_tumor_volume, on='Mouse ID')

# Calculate the correlation coefficient between mouse weight and average tumor volume
correlation_coefficient, _ = pearsonr(merged_df['Weight (g)'], merged_df['Tumor Volume (mm3)'])
print("Correlation Coefficient:", correlation_coefficient)

# Perform linear regression for mouse weight and average tumor volume
slope, intercept, r_value, p_value, std_err = linregress(merged_df['Weight (g)'], merged_df['Tumor Volume (mm3)'])

# Create a scatter plot of mouse weight vs. average observed tumor volume
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.scatter(merged_df['Weight (g)'], merged_df['Tumor Volume (mm3)'], marker='o', color='b', edgecolors='black')

# Add the linear regression line to the plot
regression_line = slope * merged_df['Weight (g)'] + intercept
plt.plot(merged_df['Weight (g)'], regression_line, color='red', label='Linear Regression')

# Set the plot title and axis labels
plt.title('Mouse Weight vs. Average Observed Tumor Volume for Capomulin Regimen')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Add a legend to the plot
plt.legend()

# Display the plot
plt.grid(True)  # Optional: Add grid lines
plt.show()
```
